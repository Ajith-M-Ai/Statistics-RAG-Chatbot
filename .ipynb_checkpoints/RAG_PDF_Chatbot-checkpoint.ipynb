# Step 5: Text Chunking

## Why Chunking is Required

Large Language Models cannot process an entire PDF at once.

To make retrieval efficient, the extracted text is divided into smaller pieces called chunks.

Benefits of Chunking:

- Reduces memory usage
- Improves retrieval accuracy
- Enables semantic search
- Improves RAG performance

In this project, RecursiveCharacterTextSplitter is used to create overlapping chunks.

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)

print("Total Chunks:", len(chunks))
```

# Understanding Chunk Size and Chunk Overlap

## Chunk Size = 500

Each chunk contains approximately 500 characters.

## Chunk Overlap = 50

The last 50 characters of one chunk are repeated in the next chunk.

This prevents loss of important context between chunks.

Example:

Chunk 1:
Machine Learning is a subset of Artificial Intelligence...

Chunk 2:
Artificial Intelligence focuses on developing systems...

The overlap helps maintain continuity.

# Step 6: Generating Embeddings

## What are Embeddings?

Embeddings are numerical vector representations of text.

They capture the semantic meaning of text and allow the system to find similar information based on meaning rather than exact keyword matching.

Example:

Text:
"Machine Learning"

Embedding:
[0.23, -0.45, 0.81, ...]

These embeddings will later be stored in a FAISS Vector Database for semantic retrieval.

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
print(type(embeddings))
```

# Creating Sample Embeddings

To verify that the embedding model is working correctly, we generate embeddings for a sample sentence.

```python
sample_embedding = embeddings.embed_query(
    "What is Machine Learning?"
)

print("Embedding Length:", len(sample_embedding))
```

# Step 7: Creating the FAISS Vector Database

## What is FAISS?

FAISS (Facebook AI Similarity Search) is a vector database used to store embeddings.

It enables fast similarity search by comparing vector representations of text.

In a RAG system:

Chunks
↓
Embeddings
↓
FAISS Database

When a user asks a question, FAISS retrieves the most relevant chunks before sending them to the LLM.

```python
from langchain_community.vectorstores import FAISS
```

```python
vectorstore = FAISS.from_documents(
    chunks,
    embeddings
)
```

# Vector Store Created

# Step 9: Creating a Retriever

A Retriever is responsible for fetching the most relevant chunks from the vector database.

Instead of manually calling similarity_search(), the retriever automates the retrieval process.

```python
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={
        "k":5,
        "fetch_k":20
    }
)
```

```python
retrieved_docs = retriever.invoke(
    "What is population?"
)
```

# Step 10: Loading FLAN-T5 Model

FLAN-T5 is a text-to-text model that can generate answers from retrieved PDF content.

We load the tokenizer and model directly instead of using the pipeline shortcut.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name
)

print("FLAN-T5 Loaded Successfully")
```

# Step 11: Testing FLAN-T5

We test whether FLAN-T5 can generate answers correctly before integrating it with the RAG pipeline.

```python
question = """
Answer the following question:

What is population?
"""

inputs = tokenizer(
    question,
    return_tensors="pt"
)

outputs = model.generate(
    **inputs,
    max_new_tokens=100
)

answer = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True
)

print(answer)
```

# Step 12: Creating RAG Question Answering Function

This function:

1. Receives a user question
2. Retrieves relevant chunks from the vector database
3. Creates a prompt with the retrieved context
4. Sends the prompt to FLAN-T5
5. Returns the generated answer

# Step 13: Testing the RAG Chatbot

The question is sent to the retriever.

Relevant chunks are fetched from the PDF.

FLAN-T5 generates an answer using the retrieved context.

```python
def ask_question(question):

    docs = retriever.invoke(question)

    context = "\n".join(
        [doc.page_content for doc in docs]
    )

    prompt = f"""
Context:
{context}

Question:
{question}

Give a complete definition in one sentence.

Answer:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=512
    )

    outputs = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=False
    )

    answer = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )

    return answer
```

```python
print(ask_question("What is population?"))
print(ask_question("What is sample?"))
print(ask_question("What is data?"))
```

# Step 14: Testing Multiple Questions

```python
questions = [
    "What is population?",
    "What is sample?",
    "What is inferential statistics?",
    "What are two kinds of populations?",
    "What is data?"
]

for q in questions:
    print("="*50)
    print("Question:", q)
    print("Answer:", ask_question(q))
    print()
```
